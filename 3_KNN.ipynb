```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the iris dataset
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.Series(iris.target, name='target')

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a k-NN classifier with k=5
knn = KNeighborsClassifier(n_neighbors=5)

# Train the classifier
knn.fit(X_train, y_train)

# Make predictions on the test set
y_pred = knn.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Print correct and wrong predictions
print("\nCorrect and Wrong Predictions:")
for i in range(len(y_test)):
    if y_test.iloc[i] == y_pred[i]:
        print(f"Instance {i + 1}: Actual = {iris.target_names[y_test.iloc[i]]}, Predicted = {iris.target_names[y_pred[i]]} (Correct)")
    else:
        print(f"Instance {i + 1}: Actual = {iris.target_names[y_test.iloc[i]]}, Predicted = {iris.target_names[y_pred[i]]} (Wrong)")

# Visualization (using sepal length and sepal width)
plt.figure(figsize=(8, 6))
plt.scatter(X_test['sepal length (cm)'], X_test['sepal width (cm)'], c=y_test, cmap='viridis', edgecolors='k', marker='o', s=80, label='Actual')
plt.scatter(X_test['sepal length (cm)'], X_test['sepal width (cm)'], c=y_pred, cmap='viridis', edgecolors='k', marker='s', s=40, label='Predicted')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('k-NN Classification (k=5)')
plt.legend()
plt.grid(True)
plt.show()
```
